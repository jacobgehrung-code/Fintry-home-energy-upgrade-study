```python
# =========================================
# STEP 0: OVERVIEW (Markdown Description)
# =========================================

from IPython.display import display, Markdown

display(Markdown("""
# ESP-r Batch Execution Script

This notebook orchestrates the sequential execution of multiple ESP-r and PV modeling scripts.

## Purpose
To automate the workflow by running all required notebooks in a fixed order, ensuring consistency and reproducibility.

## Key Features
- Executes notebooks in a defined sequence
- Stops on error for easier debugging
- Logs progress for each step
- Uses `papermill` for robust notebook execution

## Execution Order
1. Copy models to folders  
2. Generate geometry files  
3. Create parameter text files (ACH, latitude, longitude, etc.)  
4. Prepare project scripts  
5. Run simulations  
6. Process heat loads  
7. Process PV results  

## Requirements
- papermill (`pip install papermill`)
- All notebook paths correctly set

"""))
```

```python
# =========================================
# STEP 1: INSTALL & IMPORT DEPENDENCIES
# =========================================
# Description: Ensures papermill is available and imports required modules

import sys
import subprocess

def install(package):
    subprocess.check_call([sys.executable, "-m", "pip", "install", package])

try:
    import papermill as pm
except ImportError:
    install("papermill")
    import papermill as pm

import os
```

```python
# =========================================
# STEP 2: DEFINE BASE DIRECTORIES
# =========================================
# Description: Set root paths for ESP-r scripts

BASE_DIR = "/Users/jakegehrung/Desktop/data_raw/notebooks/ESP-r_scripts"

print("Base directory set to:", BASE_DIR)
```

```python
# =========================================
# STEP 3: DEFINE EXECUTION ORDER
# =========================================
# Description: Ordered list of notebooks to run

notebooks = [
    "4-copy_models_to_folders.ipynb",
    "5-2-SD2_AND_D2_geo_file.ipynb",
    "9-ach_txt_file.ipynb",
    "10-latitude_txt_file.ipynb",
    "11-longitude_txt_file.ipynb",
    "12-rotate_txt_file.ipynb",
    "13-roof_txt_file.ipynb",
    "14-walls_txt_file.ipynb",
    "15-windows_txt_file.ipynb",
    "16-prj_script_mode_all.ipynb",
    "19-pv_panel_quantity.ipynb",
    "7-simulate_all.ipynb",
    "8-heat_load_raw.ipynb",
    "6-heat_load_all.ipynb",
    "17-pv_res.ipynb",
    "18-pv_power_clean_all.ipynb"
]

print(f"{len(notebooks)} notebooks queued for execution.")
```

```python
# =========================================
# STEP 4: VALIDATE FILE EXISTENCE
# =========================================
# Description: Check all notebooks exist before execution

missing_files = []

for nb in notebooks:
    full_path = os.path.join(BASE_DIR, nb)
    if not os.path.exists(full_path):
        missing_files.append(nb)

if missing_files:
    print("Missing files detected:")
    for f in missing_files:
        print("-", f)
    raise FileNotFoundError("One or more notebooks are missing. Fix paths before running.")
else:
    print("All notebooks found successfully.")
```

```python
# =========================================
# STEP 5: EXECUTE NOTEBOOKS SEQUENTIALLY
# =========================================
# Description: Runs each notebook in order using papermill

for i, nb in enumerate(notebooks, start=1):
    input_path = os.path.join(BASE_DIR, nb)
    output_path = os.path.join(BASE_DIR, f"executed_{nb}")
    
    print(f"\n[STEP {i}] Running: {nb}")
    
    try:
        pm.execute_notebook(
            input_path=input_path,
            output_path=output_path,
            log_output=True
        )
        print(f"[STEP {i}] Completed successfully.")
        
    except Exception as e:
        print(f"[STEP {i}] FAILED.")
        print("Error:", e)
        raise  # Stop execution immediately for debugging
```

```python
# =========================================
# STEP 6: FINAL STATUS MESSAGE
# =========================================
# Description: Confirms successful completion of entire workflow

print("\nAll notebooks executed successfully. Workflow complete.")
```
