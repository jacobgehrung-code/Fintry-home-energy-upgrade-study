```python
# ============================================================
# STEP 1: Load libraries and define file paths
# Description: Imports required libraries and sets input/output file paths.
# ============================================================

import pandas as pd

input_file = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_D2_and_SD2_RESULTS.csv"
output_file = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_D2_and_SD2_RESULTS.csv"
```

```python
# ============================================================
# STEP 2: Script overview (Markdown for documentation)
# Description: Explains what this script does for future reference.
# ============================================================

from IPython.display import display, Markdown

display(Markdown("""
### Script: Add MODEL_CONSUMPTION_CURRENT Column

This script:
1. Loads an EPC dataset from a CSV file.
2. Calculates a new column:
   
   MODEL_CONSUMPTION_CURRENT = (BASELINE_ENERGY / 1000) / TOTAL_FLOOR_AREA

   - BASELINE_ENERGY is assumed to be in kWh/year.
   - Dividing by 1000 converts it to MWh/year.
   - Dividing by TOTAL_FLOOR_AREA normalises it per m².

3. Saves the updated dataset to a new CSV file.

Useful for comparing energy intensity across dwellings.
"""))
```

```python
# ============================================================
# STEP 3: Load dataset
# Description: Reads the CSV into a pandas DataFrame.
# ============================================================

df = pd.read_csv(input_file)

# Preview first few rows
df.head()
```

```python
# ============================================================
# STEP 4: Validate required columns
# Description: Ensures necessary columns exist before calculation.
# ============================================================

required_columns = ["BASELINE_ENERGY", "TOTAL_FLOOR_AREA"]

missing_cols = [col for col in required_columns if col not in df.columns]

if missing_cols:
    raise ValueError(f"Missing required columns: {missing_cols}")

print("All required columns are present.")
```

```python
# ============================================================
# STEP 5: Create MODEL_CONSUMPTION_CURRENT column
# Description: Performs the calculation and adds new column.
# ============================================================

df["MODEL_CONSUMPTION_CURRENT"] = (df["BASELINE_ENERGY"] / 1000) / df["TOTAL_FLOOR_AREA"]

# Preview results
df[["BASELINE_ENERGY", "TOTAL_FLOOR_AREA", "MODEL_CONSUMPTION_CURRENT"]].head()
```

```python
# ============================================================
# STEP 6: Handle potential divide-by-zero or invalid values
# Description: Replaces infinite values (e.g., zero floor area) with NaN.
# ============================================================

import numpy as np

df["MODEL_CONSUMPTION_CURRENT"] = df["MODEL_CONSUMPTION_CURRENT"].replace([np.inf, -np.inf], np.nan)

# Check for NaNs
print("Number of NaN values in MODEL_CONSUMPTION_CURRENT:", df["MODEL_CONSUMPTION_CURRENT"].isna().sum())
```

```python
# ============================================================
# STEP 7: Save updated dataset
# Description: Writes the updated DataFrame to a new CSV file.
# ============================================================

df.to_csv(output_file, index=False)

print(f"Updated file saved to: {output_file}")
```
