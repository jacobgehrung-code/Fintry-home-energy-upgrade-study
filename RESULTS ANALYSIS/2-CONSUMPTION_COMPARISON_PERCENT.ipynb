```python
# STEP 1: Import libraries and describe script
# ------------------------------------------
# This script:
# 1. Loads an EPC dataset from a CSV file
# 2. Creates a new column called "CONSUMPTION_COMPARISON_PERCENT"
#    using the formula:
#    [(MODEL_CONSUMPTION_CURRENT - ENERGY_CONSUMPTION_CURRENT) 
#     / ENERGY_CONSUMPTION_CURRENT] * 100
# 3. Handles potential division-by-zero safely
# 4. Saves the updated dataset to a new CSV file

import pandas as pd
import numpy as np
```

```python
# STEP 2: Load the dataset
# -----------------------
input_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_D2_and_SD2_RESULTS.csv"

df = pd.read_csv(input_path)

# Preview the first few rows
df.head()
```

```python
# STEP 3: Validate required columns exist
# --------------------------------------
required_columns = ["MODEL_CONSUMPTION_CURRENT", "ENERGY_CONSUMPTION_CURRENT"]

missing_cols = [col for col in required_columns if col not in df.columns]

if missing_cols:
    raise ValueError(f"Missing required columns: {missing_cols}")
else:
    print("All required columns are present.")
```

```python
# STEP 4: Create CONSUMPTION_COMPARISON_PERCENT column
# ---------------------------------------------------
# Avoid division by zero by replacing 0 with NaN

df["CONSUMPTION_COMPARISON_PERCENT"] = (
    (df["MODEL_CONSUMPTION_CURRENT"] - df["ENERGY_CONSUMPTION_CURRENT"]) /
    df["ENERGY_CONSUMPTION_CURRENT"].replace(0, np.nan)
) * 100

# Preview results
df[["MODEL_CONSUMPTION_CURRENT", "ENERGY_CONSUMPTION_CURRENT", "CONSUMPTION_COMPARISON_PERCENT"]].head()
```

```python
# STEP 5: Save updated dataset
# ---------------------------
output_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_D2_and_SD2_RESULTS.csv"

df.to_csv(output_path, index=False)

print(f"Updated file saved to: {output_path}")
```
