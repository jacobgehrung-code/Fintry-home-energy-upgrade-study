```python
# STEP 1 — Import Libraries
# ------------------------------------------------
# Import required modules for notebook execution

import papermill as pm
import os
```

```python
# STEP 2 — Define Base Directory
# ------------------------------------------------
# Set the directory where all notebooks are stored

BASE_DIR = "/Users/jakegehrung/Desktop/data_raw/notebooks/RESULTS ANALYSIS"
```

```python
# STEP 3 — Define Execution Order
# ------------------------------------------------
# List notebooks in the exact order required

notebooks = [
    "1-MODELED_CONSUMPTION_CURRENT.ipynb",
    "2-CONSUMPTION_COMPARISON_PERCENT.ipynb",
    "3-ECON_POS.ipynb",
    "4-BASELINE_STATS.ipynb",
    "5-HP_STATS.ipynb",
    "6-PV_STATS.ipynb"
]
```

```python
# STEP 4 — Define Output Directory (Optional but Recommended)
# ------------------------------------------------
# This stores executed notebooks for debugging and traceability

OUTPUT_DIR = os.path.join(BASE_DIR, "executed_notebooks")
os.makedirs(OUTPUT_DIR, exist_ok=True)
```

```python
# STEP 5 — Execute Notebooks Sequentially
# ------------------------------------------------
# Runs each notebook in order and stops if an error occurs

for i, nb in enumerate(notebooks, start=1):
    input_path = os.path.join(BASE_DIR, nb)
    output_path = os.path.join(OUTPUT_DIR, f"executed_{nb}")
    
    print(f"\nRunning Step {i}: {nb}")
    
    try:
        pm.execute_notebook(
            input_path=input_path,
            output_path=output_path
        )
        print(f"Completed Step {i}: {nb}")
        
    except Exception as e:
        print(f"Error in Step {i}: {nb}")
        print(str(e))
        break
```

```python
# STEP 6 — Completion Message
# ------------------------------------------------
# Confirms pipeline completion

print("\nPipeline execution finished.")
```
