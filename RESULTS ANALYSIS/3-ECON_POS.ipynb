```python
# ============================
# STEP 0: SCRIPT DESCRIPTION
# ============================
"""
This script processes an EPC-derived energy dataset for 54 dwellings in rural Scotland.

It creates binary "economically positive" indicators for each retrofit measure based on
whether Net Present Value (NPV) is greater than zero.

For each upgrade type (Windows, Roof, Walls, FAB, Heat Pump, Solar Thermal, PV, Battery),
the script generates two new sets of columns:
- *_ECON_POS_FIXED (based on fixed NPV results)
- *_ECON_POS_FLEX (based on flexible NPV results)

Output: a new CSV file with added binary economic viability indicators.
"""

# ============================
# STEP 1: LOAD DATA
# Description: Load EPC dataset and prepare environment
# ============================

import pandas as pd
import numpy as np

file_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_D2_and_SD2_RESULTS.csv"
df = pd.read_csv(file_path)

print("Data loaded successfully. Shape:", df.shape)
```

```python
# ============================
# STEP 2: CREATE ECONOMIC POSITIVE COLUMNS (FIXED NPV)
# Description: Generate binary indicators (1 if NPV > 0 else 0) for FIXED case
# ============================

fixed_npv_cols = {
    "WINDOWS": "WINDOWS_NPV_FIXED",
    "ROOF": "ROOF_NPV_FIXED",
    "WALLS": "WALLS_NPV_FIXED",
    "FAB": "FAB_NPV_FIXED",
    "HP": "HP_NPV_FIXED",
    "SOLAR_THERMAL": "SOLAR_THERMAL_NPV_FIXED",
    "PV": "PV_NPV_FIXED",
    "BATTERY": "BATTERY_NPV_FIXED"
}

for key, col in fixed_npv_cols.items():
    new_col = f"{key}_ECON_POS_FIXED"
    
    df[new_col] = np.where(
        df[col].fillna(0) > 0,
        1,
        0
    )

print("Fixed NPV economic positivity columns created.")
```

```python
# ============================
# STEP 3: CREATE ECONOMIC POSITIVE COLUMNS (FLEX NPV)
# Description: Generate binary indicators (1 if NPV > 0 else 0) for FLEX case
# ============================

flex_npv_cols = {
    "WINDOWS": "WINDOWS_NPV_FLEX",
    "ROOF": "ROOF_NPV_FLEX",
    "WALLS": "WALLS_NPV_FLEX",
    "FAB": "FAB_NPV_FLEX",
    "HP": "HP_NPV_FLEX",
    "SOLAR_THERMAL": "SOLAR_THERMAL_NPV_FLEX",
    "PV": "PV_NPV_FLEX",
    "BATTERY": "BATTERY_NPV_FLEX"
}

for key, col in flex_npv_cols.items():
    new_col = f"{key}_ECON_POS_FLEX"
    
    df[new_col] = np.where(
        df[col].fillna(0) > 0,
        1,
        0
    )

print("Flex NPV economic positivity columns created.")
```

```python
# ============================
# STEP 4: SAVE UPDATED DATASET
# Description: Export dataset with new economic viability columns
# ============================

output_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_D2_and_SD2_RESULTS.csv"

df.to_csv(output_path, index=False)

print("Updated dataset saved successfully at:")
print(output_path)
print("Final shape:", df.shape)
```
