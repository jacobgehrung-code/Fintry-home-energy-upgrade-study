# %% [Markdown]
# Step 0: Overview
# Description:
# This script automates ESP-r heat load extraction for all buildings in a dataset.
#
# Key features:
# - Iterates through all building folders in the baseline_models directory
# - Automatically detects the model folder inside each building folder
# - Navigates to the cfg directory for each building
# - Checks if 'annual.res' exists:
#     - If NOT → skips the building
# - Checks if 'heat_load_raw.csv' already exists:
#     - If YES → deletes it before running extraction
# - Runs ESP-r 'res' in script mode using the provided extraction script
# - Logs progress and skipped buildings for debugging
#
# Requirements:
# - ESP-r installed and accessible via full path (/opt/esp-r/bin/res)
# - Valid extraction script: heat_load_raw.txt
# - Folder structure must match expected format
#
# Output:
# - heat_load_raw.csv generated inside each valid cfg folder

```python
# %%
# Step 1: Import libraries and define helper function
# Description: Sets up required modules and a helper function to run terminal commands

import os
import subprocess

def run(cmd):
    print(f"\n>>> {cmd}")
    subprocess.run(cmd, shell=True, executable="/bin/bash")
```

```python
# %%
# Step 2: Define key paths and constants
# Description: Sets base directories and ESP-r executable path

base_model_path = "/Users/jakegehrung/Desktop/data_raw/baseline_models"
extract_results_path = "/Users/jakegehrung/Desktop/data_raw/ESP-r_automation_txt/heat_load_raw.txt"
res_path = "/opt/esp-r/bin/res"

res_file = "annual.res"
output_csv = "heat_load_raw.csv"
```

```python
# %%
# Step 3: Get list of all building IDs
# Description: Lists all folders in baseline_models directory (each = one building)

building_ids = [
    d for d in os.listdir(base_model_path)
    if os.path.isdir(os.path.join(base_model_path, d))
]

print(f"Found {len(building_ids)} buildings")
```

```python
# %%
# Step 4: Loop through each building and correctly locate model folder
# Description:
# - Enters each building folder
# - Searches for a valid model folder from known naming patterns
# - Navigates into model_folder/cfg
# - Performs checks and runs extraction

# Known valid model folder names
valid_model_names = {
    "SemiDetached_2F_fab", "Detached_2F_fab"
}

for building_id in building_ids:
    print(f"\n===== Processing Building: {building_id} =====")
    
    building_path = os.path.join(base_model_path, building_id)
    
    # Find the correct model folder using known names
    model_folder = None
    for d in os.listdir(building_path):
        if d in valid_model_names and os.path.isdir(os.path.join(building_path, d)):
            model_folder = d
            break
    
    if model_folder is None:
        print("No valid model folder found → Skipping")
        continue
    
    print(f"Model folder found: {model_folder}")
    
    cfg_path = os.path.join(building_path, model_folder, "cfg")
    
    if not os.path.isdir(cfg_path):
        print("No cfg folder found → Skipping")
        continue
    
    os.chdir(cfg_path)
    
    # Check for annual.res
    if not os.path.exists(res_file):
        print("annual.res not found → Skipping")
        continue
    
    # Delete existing CSV if present
    if os.path.exists(output_csv):
        print("Existing heat_load_raw.csv found → Deleting")
        os.remove(output_csv)
    
    # Run ESP-r extraction
    command = f"{res_path} -mode script -file {res_file} < {extract_results_path}"
    run(command)
    
    print(f"Completed: {building_id}")
```

```python
# %%
# Step 5: Final summary
# Description: Confirms script execution completion

print("\n=== Batch heat load extraction complete ===")
```
