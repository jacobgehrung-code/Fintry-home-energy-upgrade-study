```python
# %% STEP 1 — Overview (Markdown Description)
# This script iterates through all building folders in the baseline_models directory.
# For each building:
#   - It checks for the presence of either "SemiDetached_2F" or "Detached_2F" model folders.
#   - If neither exists, the building is skipped.
#   - If one exists, it creates 4 copies of that folder within the same building directory.
#   - The copies are renamed with suffixes: _windows, _walls, _roof, _fab
#
# IMPORTANT UPDATE:
#   - If a destination folder already exists, it will now be DELETED and REPLACED
#     with a fresh copy instead of being skipped.
#
# This ensures the script always produces clean, up-to-date copies.

import os
import shutil

baseline_models_path = "/Users/jakegehrung/Desktop/data_raw/baseline_models"
```

```python
# %% STEP 2 — Define model types and suffixes
# Defines which base folders to look for and how to rename copies.

model_types = ["SemiDetached_2F", "Detached_2F"]
suffixes = ["windows", "walls", "roof", "fab"]
```

```python
# %% STEP 3 — Identify building folders
# Collects all valid building directories in the baseline_models path.

building_folders = [
    f for f in os.listdir(baseline_models_path)
    if os.path.isdir(os.path.join(baseline_models_path, f))
]

print(f"Found {len(building_folders)} building folders")
```

```python
# %% STEP 4 — Process each building and replace existing copies
# For each building:
#   - Detect model folder
#   - Delete existing copies (if present)
#   - Create fresh copies

for building in building_folders:
    building_path = os.path.join(baseline_models_path, building)
    
    found_model = None
    
    # Check for model folder
    for model in model_types:
        model_path = os.path.join(building_path, model)
        if os.path.exists(model_path):
            found_model = model
            break
    
    if not found_model:
        print(f"Skipping {building}: No valid model folder found")
        continue
    
    print(f"Processing {building}: Found {found_model}")
    
    source_path = os.path.join(building_path, found_model)
    
    # Create (or replace) copies
    for suffix in suffixes:
        new_folder_name = f"{found_model}_{suffix}"
        destination_path = os.path.join(building_path, new_folder_name)
        
        # NEW LOGIC: Remove existing folder before copying
        if os.path.exists(destination_path):
            shutil.rmtree(destination_path)
            print(f"  Replacing existing: {new_folder_name}")
        
        shutil.copytree(source_path, destination_path)
        print(f"  Created: {new_folder_name}")
```

```python
# %% STEP 5 — Completion message
# Confirms script has finished running.

print("Processing complete.")
```
