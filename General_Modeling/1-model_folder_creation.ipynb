```python
import os
import pandas as pd

# File paths
csv_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_with_heating_code.csv"
output_dir = "/Users/jakegehrung/Desktop/data_raw/baseline_models"

# Ensure output directory exists
os.makedirs(output_dir, exist_ok=True)

# Load dataset
df = pd.read_csv(csv_path)

# Get unique building reference numbers (drop NaNs just in case)
building_ids = df["BUILDING_REFERENCE_NUMBER"].dropna().unique()

# Create folders
for bid in building_ids:
    folder_name = str(int(bid)) if isinstance(bid, float) else str(bid)
    folder_path = os.path.join(output_dir, folder_name)
    
    os.makedirs(folder_path, exist_ok=True)

print(f"Created {len(building_ids)} folders in {output_dir}")
```
