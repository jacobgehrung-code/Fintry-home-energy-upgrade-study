"""
This script processes electricity data for 117 buildings.

For each building:
1. It opens the file: energy_results.csv inside the TOTAL folder.
2. It processes ONLY the following column pair:
      electricity_tot_8 → electricity_exp_8

3. For every row:
   - If electricity_tot_8 >= 0 → electricity_exp_8 = 0
   - If electricity_tot_8 < 0:
         → electricity_exp_8 = electricity_tot_8
         → electricity_tot_8 is overwritten to 0

4. The script overwrites the original CSV file with the updated values.

The script loops through all building folders automatically.
"""

```python
# %%
# =========================================
# STEP 2: IMPORT LIBRARIES
# =========================================
import os
import pandas as pd
```

```python
# %%
# =========================================
# STEP 3: DEFINE FILE PATHS
# =========================================
baseline_models_path = "/Users/jakegehrung/Desktop/data_raw/baseline_models"
```

```python
# %%
# =========================================
# STEP 4: DEFINE TARGET COLUMNS
# =========================================
tot_col = "electricity_tot_3"
exp_col = "electricity_exp_3"
```

```python
# %%
# =========================================
# STEP 5: PROCESS EACH BUILDING
# =========================================
for building_id in os.listdir(baseline_models_path):
    building_path = os.path.join(baseline_models_path, building_id)
    
    # Ensure it's a directory
    if not os.path.isdir(building_path):
        continue
    
    total_folder = os.path.join(building_path, "TOTAL")
    file_path = os.path.join(total_folder, "energy_results.csv")
    
    # Check file exists
    if not os.path.exists(file_path):
        print(f"Skipping {building_id}: file not found")
        continue
    
    try:
        df = pd.read_csv(file_path)
        
        # Ensure exp column exists
        if exp_col not in df.columns:
            df[exp_col] = 0
        
        # Check tot column exists
        if tot_col not in df.columns:
            print(f"{building_id}: Missing column {tot_col}")
            continue
        
        # Apply logic
        negative_mask = df[tot_col] < 0
        
        df.loc[~negative_mask, exp_col] = 0
        df.loc[negative_mask, exp_col] = df.loc[negative_mask, tot_col]
        df.loc[negative_mask, tot_col] = 0
        
        # Save back to CSV
        df.to_csv(file_path, index=False)
        
        print(f"Processed building {building_id}")
    
    except Exception as e:
        print(f"Error processing {building_id}: {e}")
```

```python
# %%
# =========================================
# STEP 6: OPTIONAL — SUMMARY CHECK
# =========================================
print("Processing complete.")
```
