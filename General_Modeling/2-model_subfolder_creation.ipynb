```python
# STEP 1: Import Libraries and Define Paths
# Description: Load required libraries and define file paths for dataset and folders

import os
import shutil
import pandas as pd

# File paths
csv_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_D2_windows.csv"
base_dir = "/Users/jakegehrung/Desktop/data_raw/baseline_models"

# Folder names to create
subfolders = ["LIGHTING", "VENTILATION", "HEATING", "APPLIANCES", "HOT_WATER", "TOTAL"]
```

```python
# STEP 2: Load Dataset
# Description: Read the CSV file and extract BUILDING_REFERENCE_NUMBER column

df = pd.read_csv(csv_path)

# Ensure column exists
if "BUILDING_REFERENCE_NUMBER" not in df.columns:
    raise ValueError("Column 'BUILDING_REFERENCE_NUMBER' not found in dataset.")

# Get unique building IDs as strings
building_ids = df["BUILDING_REFERENCE_NUMBER"].dropna().astype(str).unique()

print(f"Found {len(building_ids)} unique building IDs.")
```

```python
# STEP 3: Create/Replace Subfolders
# Description: For each building folder, create required subfolders and replace if they exist

for building_id in building_ids:
    building_path = os.path.join(base_dir, building_id)
    
    if not os.path.isdir(building_path):
        print(f"Skipping missing folder: {building_path}")
        continue

    for folder in subfolders:
        target_path = os.path.join(building_path, folder)
        
        # Remove folder if it exists
        if os.path.exists(target_path):
            shutil.rmtree(target_path)
        
        # Create new folder
        os.makedirs(target_path)
    
    print(f"Processed: {building_id}")
```

```python
# STEP 4: Completion Summary
# Description: Final confirmation message

print("All folders created and replaced successfully.")
```
