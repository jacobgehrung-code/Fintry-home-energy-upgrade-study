# STEP 1 — Overview
# This script copies a set of reference CSV files into every building's TOTAL folder.
# It:
# 1. Defines the source directory containing the reference CSVs
# 2. Defines the baseline_models directory containing building folders
# 3. Iterates through each building folder
# 4. Locates (or checks for) the TOTAL subfolder
# 5. Copies all required CSV files into each TOTAL folder
# 6. Logs progress and any missing folders/files for troubleshooting

```python
# STEP 2 — Import libraries and define paths

import os
import shutil

# Path to reference CSV files
reference_dir = "/Users/jakegehrung/Desktop/data_raw/REFERENCE FILES"

# Path to building model folders
baseline_models_dir = "/Users/jakegehrung/Desktop/data_raw/baseline_models"

# List of files to copy
files_to_copy = [
    "annual_aggregate.csv",
    "annual_savings.csv",
    "annual_results.csv",
    "carbon_results.csv",
    "fixed_cost_results.csv",
    "flex_cost_results.csv",
    "energy_results.csv",
    "energy_results_tot.csv"
]

print("Paths and file list defined.")
```

```python
# STEP 3 — Validate reference files exist

missing_files = []

for file_name in files_to_copy:
    file_path = os.path.join(reference_dir, file_name)
    if not os.path.exists(file_path):
        missing_files.append(file_name)

if missing_files:
    print("ERROR: The following reference files are missing:")
    for f in missing_files:
        print(f" - {f}")
else:
    print("All reference files found.")
```

```python
# STEP 4 — Copy files into each TOTAL folder

copied_count = 0
skipped_count = 0

for building_id in os.listdir(baseline_models_dir):
    building_path = os.path.join(baseline_models_dir, building_id)

    # Ensure it's a directory
    if not os.path.isdir(building_path):
        continue

    total_folder = os.path.join(building_path, "TOTAL")

    # Check if TOTAL folder exists
    if not os.path.exists(total_folder):
        print(f"Skipping {building_id} — TOTAL folder not found")
        skipped_count += 1
        continue

    # Copy each file
    for file_name in files_to_copy:
        src = os.path.join(reference_dir, file_name)
        dst = os.path.join(total_folder, file_name)

        shutil.copy2(src, dst)

    copied_count += 1

print("\nCopy process complete.")
print(f"Folders updated: {copied_count}")
print(f"Folders skipped (no TOTAL): {skipped_count}")
```
