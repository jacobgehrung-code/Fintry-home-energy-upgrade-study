```python
# Step 1: Description
# ---------------------------------------
# This script:
# 1. Loads an EPC dataset from a specified CSV file
# 2. Filters the dataset to include only rows where the "ARCHETYPE" column
#    is either "SemiDetached_2F" or "Detached_2F"
# 3. Saves the filtered dataset to a new CSV file at the specified output path
#
# Designed for use in Jupyter Notebook (Mac / VSCode compatible)

import pandas as pd
```

```python
# Step 2: Load the dataset
# ---------------------------------------
# Reads the CSV file into a pandas DataFrame

input_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_NPV_IRR.csv"

df = pd.read_csv(input_path)

# Preview to confirm successful load
df.head()
```

```python
# Step 3: Filter rows based on ARCHETYPE
# ---------------------------------------
# Keeps only rows where ARCHETYPE is one of the specified values

valid_archetypes = ["SemiDetached_2F", "Detached_2F"]

filtered_df = df[df["ARCHETYPE"].isin(valid_archetypes)]

# Check result
filtered_df["ARCHETYPE"].value_counts()
```

```python
# Step 4: Save filtered dataset
# ---------------------------------------
# Writes the filtered DataFrame to a new CSV file

output_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_D2_and_SD2_RESULTS.csv"

filtered_df.to_csv(output_path, index=False)

print(f"Filtered dataset saved to: {output_path}")
```
