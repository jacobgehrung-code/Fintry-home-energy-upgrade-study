# STEP 0 — Script Overview (Markdown Description)
# ------------------------------------------------
# This notebook orchestrates the execution of multiple Jupyter notebooks
# in a defined sequence for your energy modelling workflow.
#
# It uses the `papermill` library to execute each notebook in order,
# ensuring dependencies between steps are respected.
#
# Features:
# - Sequential execution of notebooks
# - Clear step logging for debugging
# - Stops execution if a notebook fails
# - Saves executed copies (optional) for traceability
#
# Requirements:
# pip install papermill
#
# Directory Assumption:
# All notebooks are located in:
# /Users/jakegehrung/Desktop/data_raw/notebooks/Python_Modeling
#
# You can modify paths or execution behavior as needed.

```python
# STEP 1 — Import Libraries
# ------------------------------------------------
# Import required modules for notebook execution

import papermill as pm
import os
```

```python
# STEP 2 — Define Base Directory
# ------------------------------------------------
# Set the directory where all notebooks are stored

BASE_DIR = "/Users/jakegehrung/Desktop/data_raw/notebooks/Python_Modeling"
```

```python
# STEP 3 — Define Execution Order
# ------------------------------------------------
# List notebooks in the exact order required

notebooks = [
    "3-lighting_schedules.ipynb",
    "3-secondary_heating_activation.ipynb",
    "4-lighting_power.ipynb",
    "5-appliance_power_w_lighting.ipynb",
    "6-appliance_power.ipynb",
    "7-heat_efficiency.ipynb",
    "8-copy_second_heat_actv.ipynb",
    "9-heat_load_dis.ipynb",
    "10-heat_by_fuel.ipynb",
    "11-ventilation_fan_power.ipynb",
    "13-hot_water_efficiency.ipynb",
    "16-hot_water_load.ipynb",
    "14-hot_water_by_fuel.ipynb"
]
```

```python
# STEP 4 — Define Output Directory (Optional but Recommended)
# ------------------------------------------------
# This stores executed notebooks for debugging and traceability

OUTPUT_DIR = os.path.join(BASE_DIR, "executed_notebooks")
os.makedirs(OUTPUT_DIR, exist_ok=True)
```

```python
# STEP 5 — Execute Notebooks Sequentially
# ------------------------------------------------
# Runs each notebook in order and stops if an error occurs

for i, nb in enumerate(notebooks, start=1):
    input_path = os.path.join(BASE_DIR, nb)
    output_path = os.path.join(OUTPUT_DIR, f"executed_{nb}")
    
    print(f"\nRunning Step {i}: {nb}")
    
    try:
        pm.execute_notebook(
            input_path=input_path,
            output_path=output_path
        )
        print(f"Completed Step {i}: {nb}")
        
    except Exception as e:
        print(f"Error in Step {i}: {nb}")
        print(str(e))
        break
```

```python
# STEP 6 — Completion Message
# ------------------------------------------------
# Confirms pipeline completion

print("\nPipeline execution finished.")
```
