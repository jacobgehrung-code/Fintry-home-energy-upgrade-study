# Step 1: Overview
# --------------------------------------------------
# This script updates the "electricity_app_light_vent" column in every
# energy_results.csv file across all building folders.
#
# For each building:
# - Navigate to: baseline_models/{BUILDING_REFERENCE_NUMBER}/TOTAL/
# - Open energy_results.csv
# - Compute:
#     electricity_app_light_vent = electricity_appliances
#                                  + electricity_lighting
#                                  + electricity_ventilation
# - Save the updated CSV (overwrite existing file)
#
# The script:
# - Iterates through all building folders
# - Handles missing files or columns gracefully
# - Prints progress and any issues encountered

```python
# Step 2: Import libraries and define paths
# --------------------------------------------------
import os
import pandas as pd

# Base directory containing all building folders
BASELINE_MODELS_PATH = "/Users/jakegehrung/Desktop/data_raw/baseline_models"
```

```python
# Step 3: Define helper function to process a single file
# --------------------------------------------------
def process_energy_results(file_path):
    try:
        df = pd.read_csv(file_path)

        required_cols = [
            "electricity_appliances",
            "electricity_lighting",
            "electricity_ventilation"
        ]

        # Check required columns exist
        if not all(col in df.columns for col in required_cols):
            print(f"Missing columns in: {file_path}")
            return

        # Create or overwrite the target column
        df["electricity_app_light_vent"] = (
            df["electricity_appliances"].fillna(0) +
            df["electricity_lighting"].fillna(0) +
            df["electricity_ventilation"].fillna(0)
        )

        # Save back to the same file
        df.to_csv(file_path, index=False)

        print(f"Updated: {file_path}")

    except Exception as e:
        print(f"Error processing {file_path}: {e}")
```

```python
# Step 4: Iterate through all building folders
# --------------------------------------------------
for building_id in os.listdir(BASELINE_MODELS_PATH):
    building_path = os.path.join(BASELINE_MODELS_PATH, building_id)

    # Ensure it's a directory
    if not os.path.isdir(building_path):
        continue

    total_folder = os.path.join(building_path, "TOTAL")
    csv_path = os.path.join(total_folder, "energy_results.csv")

    if os.path.exists(csv_path):
        process_energy_results(csv_path)
    else:
        print(f"No energy_results.csv found for building: {building_id}")
```
