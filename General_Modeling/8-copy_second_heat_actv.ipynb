```python
# Step 1: Setup paths and import libraries
# Description: Define all file paths and import required modules

import os
import shutil
import pandas as pd

# File paths
epc_file = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_NPV_IRR_update.csv"
baseline_models_dir = "/Users/jakegehrung/Desktop/data_raw/baseline_models"
source_file = "/Users/jakegehrung/Desktop/data_raw/data_processed/secondary_heating_activation.csv"
```

```python
# Step 2: Load EPC dataset and extract building IDs
# Description: Read the EPC CSV and get the list of BUILDING_REFERENCE_NUMBER values

epc_df = pd.read_csv(epc_file)

# Ensure IDs are strings (important for folder matching)
building_ids = epc_df["BUILDING_REFERENCE_NUMBER"].astype(str).tolist()

print(f"Total buildings found: {len(building_ids)}")
```

```python
# Step 3: Copy file into each HEATING folder
# Description: Loop through each building ID, locate its HEATING folder, and copy the file

copied = 0
missing_folders = []
missing_heating = []

for b_id in building_ids:
    model_path = os.path.join(baseline_models_dir, b_id)
    heating_path = os.path.join(model_path, "HEATING")
    
    if not os.path.exists(model_path):
        missing_folders.append(b_id)
        continue
    
    if not os.path.exists(heating_path):
        missing_heating.append(b_id)
        continue
    
    destination_file = os.path.join(heating_path, "secondary_heating_activation.csv")
    
    try:
        shutil.copy(source_file, destination_file)
        copied += 1
    except Exception as e:
        print(f"Error copying for {b_id}: {e}")

print(f"\nFiles successfully copied: {copied}")
print(f"Missing model folders: {len(missing_folders)}")
print(f"Missing HEATING folders: {len(missing_heating)}")
```

```python
# Step 4: Optional - Inspect missing cases
# Description: Print out any IDs where folders were missing for debugging

print("\nMissing model folders:")
print(missing_folders)

print("\nMissing HEATING folders:")
print(missing_heating)
```
