```python
# STEP 1: IMPORT LIBRARIES & SCRIPT OVERVIEW
# -----------------------------------------
# Description:
# This script loads an EPC dataset, calculates PV capital cost based on
# MODELED_PV and PV_UPGRADE columns, and creates (or replaces) a column
# called HP_CAP_COST with the computed values.

import pandas as pd

# File path to dataset
file_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_windows_upgrade_types.csv"
```

```python
# STEP 2: LOAD DATASET
# --------------------
# Description:
# Reads the CSV file into a pandas DataFrame.

df = pd.read_csv(file_path)

# Preview data
df.head()
```

```python
# STEP 3: DEFINE PARAMETERS
# ------------------------
# Description:
# Sets cost parameters used in the PV capital cost calculation.

material_cost = 893
labor_cost = 107
```

```python
# STEP 4: CALCULATE HP_CAP_COST
# -----------------------------
# Description:
# Computes PV capital cost using:
# [(MODELED_PV * labor_cost) + (MODELED_PV * material_cost)] * PV_UPGRADE
# and assigns result to HP_CAP_COST column.

df["PV_CAP_COST"] = (
    (df["MODELED_PV"] * labor_cost) +
    (df["MODELED_PV"] * material_cost)
) * df["PV_UPGRADE"]
```

```python
# STEP 5: VERIFY RESULTS
# ---------------------
# Description:
# Displays key columns to confirm calculation worked correctly.

df[["MODELED_PV", "PV_UPGRADE", "PV_CAP_COST"]].head()
```

```python
# STEP 6: SAVE UPDATED DATASET
# ---------------------------
# Description:
# Writes the updated DataFrame back to CSV (overwrites original file).

df.to_csv(file_path, index=False)
```
