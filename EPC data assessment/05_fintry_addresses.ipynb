```python
# %%
# ==========================
# Block 1: Load EPC data
# ==========================

import pandas as pd

# Path to processed EPC data
input_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_processed.csv"

df = pd.read_csv(input_path)

df.shape
```

```python
# %%
# ==========================
# Block 2: Clean address components
# ==========================

address_cols = ["ADDRESS1", "ADDRESS2", "ADDRESS3", "POSTCODE"]

for col in address_cols:
    df[col] = (
        df[col]
        .fillna("")
        .astype(str)
        .str.strip()
        .str.replace(r"\s+", " ", regex=True)
    )

df[address_cols].head()
```

```python
# %%
# ==========================
# Block 3: Build FullAddress for geocoding
# ==========================

def build_full_address(row):
    parts = [
        row["ADDRESS1"],
        row["ADDRESS2"],
        row["ADDRESS3"],
        row["POSTCODE"],
        "UK"
    ]
    # Keep only non-empty parts
    return ", ".join([p for p in parts if p])

df["FullAddress"] = df.apply(build_full_address, axis=1)

df[["FullAddress"]].head(10)
```

```python
# %%
# ==========================
# Block 4: Save fintry_addresses.csv
# ==========================

output_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_addresses.csv"

address_df = df[["FullAddress"]].drop_duplicates().reset_index(drop=True)

address_df.to_csv(output_path, index=False)

output_path, address_df.shape
```
