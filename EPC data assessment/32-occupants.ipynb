```python
# STEP 1 — Load EPC dataset
# Description: Import libraries and load the EPC CSV file into a pandas DataFrame

import pandas as pd
import numpy as np

file_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_lighting_watts.csv"

df = pd.read_csv(file_path)

df.head()
```

```python
# STEP 2 — Calculate number of occupants (continuous)
# Description: Apply EPC occupancy formula using TOTAL_FLOOR_AREA (TFA)

TFA = df["TOTAL_FLOOR_AREA"]

df["NUMBER_OF_OCCUPANTS"] = (
    1
    + (
        1.76 * (1 - np.exp(-0.000349 * (TFA - 13.9) ** 2))
        + 0.0013 * (TFA - 13.9)
    )
)

df[["TOTAL_FLOOR_AREA", "NUMBER_OF_OCCUPANTS"]].head()
```

```python
# STEP 3 — Round occupants up to nearest whole number
# Description: Create OCCUPANTS_ROUNDED_UP using ceiling function

df["OCCUPANTS_ROUNDED_UP"] = np.ceil(df["NUMBER_OF_OCCUPANTS"]).astype(int)

df[["NUMBER_OF_OCCUPANTS", "OCCUPANTS_ROUNDED_UP"]].head()
```

```python
# STEP 4 — Save output dataset
# Description: Export updated EPC dataset with occupancy columns to CSV

output_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_number_of_occ.csv"

df.to_csv(output_path, index=False)

print("File saved to:", output_path)
```
