```python
# STEP 1: Load Libraries and Dataset
# Description: Import pandas and load the EPC dataset from the specified file path.

import pandas as pd

file_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_NPV_IRR_update.csv"
df = pd.read_csv(file_path)

# Preview data
df.head()
```

```python
# STEP 2: Inspect Relevant Columns
# Description: Check unique values in PROPERTY_TYPE and ARCHETYPE to understand current structure.

print("Unique PROPERTY_TYPE values:")
print(df["PROPERTY_TYPE"].unique())

print("\nUnique ARCHETYPE values (before update):")
print(df["ARCHETYPE"].unique())
```

```python
# STEP 3: Update ARCHETYPE for Flats
# Description: Replace ARCHETYPE with "Flat" wherever PROPERTY_TYPE is "Flat".

df.loc[df["PROPERTY_TYPE"].str.strip().str.lower() == "flat", "ARCHETYPE"] = "Flat"

# Verify changes
print("\nUnique ARCHETYPE values (after update):")
print(df["ARCHETYPE"].unique())
```

```python
# STEP 4: Save Updated Dataset
# Description: Save the modified dataframe back to file (overwrite or create new file).

output_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_NPV_IRR_update.csv"
df.to_csv(output_path, index=False)

print(f"Updated file saved to: {output_path}")
```
