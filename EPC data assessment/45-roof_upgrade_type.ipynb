```python
# Step 1: Load Dataset
# Description: Import required libraries and load the EPC dataset from the specified file path.

"""
This script assigns a ROOF_UPGRADE_TYPE to each dwelling in the EPC dataset
based on its ROOF_CONS value.

Steps:
1. Load the dataset
2. Define mapping logic (ROOF_CONS → ROOF_UPGRADE_TYPE)
3. Apply mapping to create/update ROOF_UPGRADE_TYPE column
4. Check for unmapped values
5. Save updated dataset
"""

import pandas as pd

# File path
file_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_wall_upgrade_types.csv"

# Load dataset
df = pd.read_csv(file_path)

# Preview
df.head()
```

```python
# Step 2: Define Mapping Dictionary
# Description: Create a dictionary that maps ROOF_CONS values to ROOF_UPGRADE_TYPE.

roof_mapping = {
    "insulated_roof_0.15": "loft",  
    "insulated_roof_0.35": "loft",  
    "insulated_roof_100mm": "loft",  
    "insulated_roof_12mm": "loft",  
    "insulated_roof_150mm": "loft",  
    "insulated_roof_200mm": "loft",  
    "insulated_roof_250mm": "loft",  
    "insulated_roof_270mm": "loft",  
    "insulated_roof_300mm": "loft",  
    "insulated_roof_350mm": "loft",  
    "insulated_roof_400mm": "loft",  
    "insulated_roof_50mm": "loft",  
    "insulated_roof_75mm": "loft",  
    "roof_insulated_rafters": "rafters",  
    "uninsulated_roof": "loft"  
}
```

```python
# Step 3: Apply Mapping
# Description: Use the mapping dictionary to populate the WALL_UPGRADE_TYPE column.

df["ROOF_UPGRADE_TYPE"] = df["ROOF_CONS"].map(roof_mapping)

# Preview results
df[["ROOF_CONS", "ROOF_UPGRADE_TYPE"]].head(10)
```

```python
# Step 4: Check for Unmapped Values
# Description: Identify any ROOF_CONS values that were not found in the mapping.

unmapped = df[df["ROOF_UPGRADE_TYPE"].isna()]["ROOF_CONS"].unique()

print("Unmapped ROOF_CONS values:")
print(unmapped)
```

```python
# Step 5: Save Updated Dataset
# Description: Save the updated dataset with ROOF_UPGRADE_TYPE filled.

output_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_roof_upgrade_types.csv"

df.to_csv(output_path, index=False)

print(f"Updated dataset saved to: {output_path}")
```
