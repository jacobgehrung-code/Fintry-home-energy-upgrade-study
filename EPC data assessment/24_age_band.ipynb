```python
import pandas as pd

# Input file path
input_file = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_cons_update.csv"

# Output file path
output_file = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_cons_update2.csv"

# Load dataset
df = pd.read_csv(input_file)

# Mapping dictionary
age_band_map = {
    "before 1919": "A",
    "1919-1929": "B",
    "1930-1949": "C",
    "1950-1964": "D",
    "1965-1975": "E",
    "1976-1983": "F",
    "1984-1991": "G",
    "1992-1998": "H",
    "1999-2002": "I",
    "2003-2007": "J",
    "2008-2011": "K",
    "2008 onwards": "K",
    "2012 onwards": "L"
}

# Create AGE_BAND column
df["AGE_BAND"] = df["CONSTRUCTION_AGE_BAND"].map(age_band_map)

# Save new CSV
df.to_csv(output_file, index=False)

print("New file created:", output_file)
```
