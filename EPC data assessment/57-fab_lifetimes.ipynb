# Step 1: Overview — Script Description
# ------------------------------------
# This script processes an EPC dataset and assigns lifetime values for building fabric elements
# (walls, windows, roof, and overall fabric) based on predefined upgrade types.
#
# INPUT:
# - CSV file containing upgrade type columns:
#   WALLS_UPGRADE_TYPE, WINDOWS_UPGRADE_TYPE, ROOF_UPGRADE_TYPE
#
# OUTPUT:
# - Updated CSV file with new/updated columns:
#   WALLS_LIFETIME, WINDOWS_LIFETIME, ROOF_LIFETIME, FAB_LIFETIME
#
# LOGIC:
# - WALLS_LIFETIME is assigned based on WALLS_UPGRADE_TYPE mapping
# - WINDOWS_LIFETIME is assigned based on WINDOWS_UPGRADE_TYPE mapping
# - ROOF_LIFETIME is assigned based on ROOF_UPGRADE_TYPE mapping
# - FAB_LIFETIME is set to a constant value (50 years) for all rows
#
# NOTES:
# - Any upgrade types not found in the mapping will result in NaN values
# - Ensure column names match exactly (case-sensitive)

```python
# Step 2: Import Libraries
# -----------------------
import pandas as pd
```

```python
# Step 3: Load Dataset
# -------------------
file_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_with_costs_2.csv"
df = pd.read_csv(file_path)

# Preview dataset
df.head()
```

```python
# Step 4: Define Lifetime Mappings
# --------------------------------

# Walls lifetime mapping
walls_lifetime_map = {
    "none": 0,
    "external": 50,
    "external_and_cavity": 30,
    "cavity": 30,
    "internal": 80
}

# Windows lifetime mapping
windows_lifetime_map = {
    "uPVC": 35
}

# Roof lifetime mapping
roof_lifetime_map = {
    "loft": 80,
    "rafters": 80
}
```

```python
# Step 5: Apply Lifetime Mappings
# ------------------------------

# Assign lifetimes based on upgrade type
df["WALLS_LIFETIME"] = df["WALLS_UPGRADE_TYPE"].map(walls_lifetime_map)
df["WINDOWS_LIFETIME"] = df["WINDOWS_UPGRADE_TYPE"].map(windows_lifetime_map)
df["ROOF_LIFETIME"] = df["ROOF_UPGRADE_TYPE"].map(roof_lifetime_map)
```

```python
# Step 6: Assign FAB_LIFETIME (Constant)
# -------------------------------------
df["FAB_LIFETIME"] = 50
```

```python
# Step 7: Validate Results
# -----------------------
# Check for missing values (unmapped categories)
print("Missing WALLS_LIFETIME values:", df["WALLS_LIFETIME"].isna().sum())
print("Missing WINDOWS_LIFETIME values:", df["WINDOWS_LIFETIME"].isna().sum())
print("Missing ROOF_LIFETIME values:", df["ROOF_LIFETIME"].isna().sum())

# Preview updated dataframe
df.head()
```

```python
# Step 8: Save Output File
# -----------------------
output_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_with_lifetimes_2.csv"
df.to_csv(output_path, index=False)

print(f"File successfully saved to: {output_path}")
```
