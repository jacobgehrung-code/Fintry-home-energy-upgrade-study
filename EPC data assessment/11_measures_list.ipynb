```python
import pandas as pd
import ast

# File paths
input_csv = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_step1_improvement_metrics.csv"
output_csv = "/Users/jakegehrung/Desktop/data_raw/data_processed/measures_list.csv"

# Load CSV
df = pd.read_csv(input_csv)

# Container for unique descriptions
unique_descriptions = set()

# Iterate through the column
for cell in df["IMPROVEMENTS_PARSED"].dropna():
    try:
        records = ast.literal_eval(cell)  # safely parse string to Python object
        for record in records:
            desc = record.get("description")
            if desc:  # filters out None and empty strings
                unique_descriptions.add(desc)
    except (ValueError, SyntaxError):
        # Skip malformed rows gracefully
        continue

# Create output DataFrame
out_df = pd.DataFrame(
    sorted(unique_descriptions),
    columns=["description"]
)

# Save to CSV
out_df.to_csv(output_csv, index=False)

print(f"Saved {len(out_df)} unique descriptions to:")
print(output_csv)
```
