```python
# Step 1: Load Dataset
# Description: Import required libraries and load the EPC dataset from the specified file path.

"""
This script assigns a WINDOWS_UPGRADE_TYPE to each dwelling in the EPC dataset
based on its WINDOW_CONS value.

Steps:
1. Load the dataset
2. Define mapping logic (WINDOW_CONS → WINDOWS_UPGRADE_TYPE)
3. Apply mapping to create/update WINDOWS_UPGRADE_TYPE column
4. Check for unmapped values
5. Save updated dataset
"""

import pandas as pd

# File path
file_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_roof_upgrade_types.csv"

# Load dataset
df = pd.read_csv(file_path)

# Preview
df.head()
```

```python
# Step 2: Define Mapping Dictionary
# Description: Create a dictionary that maps WINDOWS_CONS values to WINDOWS_UPGRADE_TYPE.

window_mapping = {
    "dbl_glz_pre_2003": "uPVC",  
    "triple_glz": "uPVC",  
    "dbl_glz_post_2003": "uPVC",  
    "single_glz": "uPVC"  
}
```

```python
# Step 3: Apply Mapping
# Description: Use the mapping dictionary to populate the WINDOWS_UPGRADE_TYPE column.

df["WINDOWS_UPGRADE_TYPE"] = df["WINDOW_CONS"].map(window_mapping)

# Preview results
df[["WINDOW_CONS", "WINDOWS_UPGRADE_TYPE"]].head(10)
```

```python
# Step 4: Check for Unmapped Values
# Description: Identify any WINDOWS_CONS values that were not found in the mapping.

unmapped = df[df["WINDOWS_UPGRADE_TYPE"].isna()]["WINDOW_CONS"].unique()

print("Unmapped WINDOW_CONS values:")
print(unmapped)
```

```python
# Step 5: Save Updated Dataset
# Description: Save the updated dataset with WINDOWS_UPGRADE_TYPE filled.

output_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_windows_upgrade_types.csv"

df.to_csv(output_path, index=False)

print(f"Updated dataset saved to: {output_path}")
```
