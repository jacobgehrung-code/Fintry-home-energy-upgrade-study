```python
# ==========================
# Block: Google Maps Geocoding (Hard-coded paths)
# ==========================

import googlemaps
import pandas as pd
import time
from pathlib import Path

# --------------------------
# Hard-coded file paths
# --------------------------
INPUT_CSV = Path("/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_addresses.csv")
OUTPUT_CSV = Path("/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_geocoded.csv")

# --------------------------
# Google Maps client
# --------------------------
gmaps = googlemaps.Client(key=MY_KEY)

# --------------------------
# Load addresses
# --------------------------
df = pd.read_csv(INPUT_CSV)  # must contain 'FullAddress'

# Add columns for coordinates
df["Latitude"] = None
df["Longitude"] = None

# --------------------------
# Geocode loop
# --------------------------
for i, address in df["FullAddress"].items():
    try:
        result = gmaps.geocode(address)

        if result:
            location = result[0]["geometry"]["location"]
            df.at[i, "Latitude"] = location["lat"]
            df.at[i, "Longitude"] = location["lng"]
        else:
            print(f"Address not found: {address}")

    except Exception as e:
        print(f"Error geocoding {address}: {e}")

    time.sleep(0.1)  # safe for ~136 addresses

# --------------------------
# Save output
# --------------------------
df.to_csv(OUTPUT_CSV, index=False)

print(f"Geocoding complete. Saved to: {OUTPUT_CSV}")
```
