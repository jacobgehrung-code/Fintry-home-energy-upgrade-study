```python
# STEP 1: Import libraries and define file paths
# Description: Load required libraries and specify input/output file paths

import pandas as pd

input_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_with_heat_codes.csv"
output_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_with_fuel_codes.csv"
```

```python
# STEP 2: Load dataset
# Description: Read the EPC dataset into a pandas DataFrame

df = pd.read_csv(input_path)

# Quick check
print(df.head())
print(df.columns)
```

```python
# STEP 3: Define heat-to-fuel mapping dictionary
# Description: Create a dictionary mapping heating system codes to fuel types

heat_to_fuel = {
    "air_source_heat_pump_radiators_electric": "electricity",
    "air_source_heat_pump_underfloor_electric": "electricity",
    "boiler_radiators_lpg": "lpg",
    "boiler_radiators_dual_fuel_mineral_wood": "mineral_wood",
    "boiler_radiators_electric": "electricity",
    "boiler_radiators_mains_gas": "mains_gas",
    "boiler_radiators_oil": "oil",
    "boiler_radiators_wood_logs": "wood_logs",
    "boiler_radiators_wood_pellets": "wood_pellets",
    "community_scheme": "wood_chips",
    "electric_storage_heaters": "electricity",
    "ground_source_heat_pump_radiators_electric": "electricity",
    "ground_source_heat_pump_underfloor_electric": "electricity",
    "no_system_present_electric_heaters_assumed": "electricity",
    "room_heaters_dual_fuel_mineral_wood": "mineral_wood",
    "room_heaters_electric": "electricity",
    "room_heaters_wood_logs": "wood_logs",
    "portable_electric_heaters": "electricity",
    "room_heaters_dual_fuel": "mineral_wood",
    "room_heaters_coal": "coal",
    "room_heaters_wood_pellets": "wood_pellets"
}
```

```python
# STEP 4: Create MAIN_FUEL_CODE column
# Description: Map main_heat values to fuel types

df["MAIN_FUEL_CODE"] = df["main_heat"].map(heat_to_fuel)

# Check for unmapped values
print("Unmapped main_heat values:")
print(df[df["MAIN_FUEL_CODE"].isna()]["main_heat"].unique())
```

```python
# STEP 5: Create SECOND_FUEL_CODE column
# Description: Map secondary_heat values to fuel types

df["SECOND_FUEL_CODE"] = df["secondary_heat"].map(heat_to_fuel)

# Check for unmapped values
print("Unmapped secondary_heat values:")
print(df[df["SECOND_FUEL_CODE"].isna()]["secondary_heat"].unique())
```

```python
# STEP 6: Save updated dataset
# Description: Export the DataFrame with new fuel code columns to CSV

df.to_csv(output_path, index=False)

print(f"File saved to: {output_path}")
```

```python
# STEP 7: Final verification
# Description: Display summary of fuel code columns for sanity check

print(df[["main_heat", "MAIN_FUEL_CODE", "secondary_heat", "SECOND_FUEL_CODE"]].head())
print("\nFuel code counts (main):")
print(df["MAIN_FUEL_CODE"].value_counts(dropna=False))
print("\nFuel code counts (secondary):")
print(df["SECOND_FUEL_CODE"].value_counts(dropna=False))
```
