What this code does
Loads the dataset from Step 8 with priority indices.
Identifies key fields that are critical for energy, carbon, financial, and building characteristics.
Counts missing values for each key field and calculates the percentage missing.
Flags fields with high missingness (>5% by default) to indicate gaps needing attention.
Marks dwellings with uncertainty in priority index due to missing values.
Exports a CSV report summarizing missing data and uncertainty for FDT.


```python
# Step 9: Missing Data / Sensitivity Reporting
# Description: Analyze the dataset to identify missing values, gaps in key variables, and potential sources of uncertainty.
# Dependencies: Requires the dataset processed through Step 8 (Priority Index calculated).

import pandas as pd
import numpy as np

# ---------------------------------------
# Step 9a: Load dataset
# Description: Load the processed dataset containing priority indices.
# File path provided by user
file_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_priority_index_step8.csv"
df = pd.read_csv(file_path)

# Quick check of data shape
print(f"Dataset contains {df.shape[0]} rows and {df.shape[1]} columns.")

# ---------------------------------------
# Step 9b: Identify key fields to check
# Description: Specify which fields are essential for prioritization and analysis
key_fields = [
    'DELTA_ENERGY_EFFICIENCY',
    'DELTA_ENVIRONMENT_IMPACT',
    'DELTA_ENERGY_CONSUMPTION',
    'PCT_ENERGY_REDUCTION',
    'DELTA_CO2_EMISSIONS',
    'PCT_CO2_REDUCTION',
    'ANNUAL_ENERGY_COST_SAVINGS',
    'NPV',
    'IRR',
    'COST_PER_KWH_REDUCED',
    'COST_PER_TON_CO2_REDUCED',
    'PRIORITY_INDEX',
    'MAIN_HEATING_CATEGORY',
    'MAIN_FUEL',
    'MAIN_HEATING_CONTROLS',
    'NUMBER_HEATED_ROOMS',
    'FLOOR_LEVEL',
    'TENURE'
]

# ---------------------------------------
# Step 9c: Count missing values in key fields
# Description: Produce a summary of missing values for each key field
missing_summary = df[key_fields].isnull().sum().reset_index()
missing_summary.columns = ['Field', 'Missing_Count']
missing_summary['Missing_Percent'] = missing_summary['Missing_Count'] / len(df) * 100

# Display summary table
print("\nMissing Data Summary:")
print(missing_summary.sort_values(by='Missing_Percent', ascending=False))

# ---------------------------------------
# Step 9d: Highlight fields with highest missingness
# Description: Identify fields that may require additional data collection or could affect prioritization
high_missing = missing_summary[missing_summary['Missing_Percent'] > 5]  # Threshold can be adjusted
print("\nFields with >5% missing values (potential gaps for modeling):")
print(high_missing)

# ---------------------------------------
# Step 9e: Sensitivity / uncertainty considerations
# Description: Identify fields that, if missing or uncertain, could impact the Priority Index or decision-making
# We create a simple flag for any dwelling with missing values in critical fields
df['PRIORITY_INDEX_UNCERTAINTY_FLAG'] = df[key_fields].isnull().any(axis=1)

num_uncertain = df['PRIORITY_INDEX_UNCERTAINTY_FLAG'].sum()
print(f"\nNumber of dwellings with potential uncertainty affecting prioritization: {num_uncertain}")

# ---------------------------------------
# Step 9f: Generate report for FDT
# Description: Save a CSV summary table for reporting purposes
report_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_missing_data_report_step9.csv"

# Combine missing summary with uncertainty flag info
report_df = missing_summary.copy()
report_df['Potential_Uncertainty_Affected'] = num_uncertain

report_df.to_csv(report_path, index=False)
print(f"\nMissing data and uncertainty report saved to: {report_path}")
```
