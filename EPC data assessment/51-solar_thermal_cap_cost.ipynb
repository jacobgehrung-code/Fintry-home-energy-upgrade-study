# STEP 1: Overview - Script Description
# -----------------------------------
# This script performs the following steps:
# 1. Loads the EPC dataset (fintry_epc_cap_cost_manual.csv).
# 2. Loads a reference CSV (solar_thermal_cap_cost.csv) containing a single value.
# 3. Extracts that value.
# 4. Creates (or replaces if already exists) a column called 'SOLAR_THERMAL_CAP_COST'
#    in the EPC dataset and fills it with the extracted value for all rows.
# 5. Saves the updated dataset back to disk (option to overwrite or create a new file).

```python
# STEP 2: Import Required Libraries
# --------------------------------
import pandas as pd
```

```python
# STEP 3: Define File Paths
# ------------------------
epc_file_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_cap_cost.csv"
solar_cost_file_path = "/Users/jakegehrung/Desktop/data_raw/REFERENCE FILES/solar_thermal_cap_cost.csv"
```

```python
# STEP 4: Load the Datasets
# -------------------------
epc_df = pd.read_csv(epc_file_path)
solar_cost_df = pd.read_csv(solar_cost_file_path)

# Preview to verify loading (optional)
print(epc_df.head())
print(solar_cost_df.head())
```

```python
# STEP 5 (FINAL FIX): Correctly Read Single-Value CSV
# --------------------------------------------------
# Your file is being interpreted as a HEADER (column name) instead of data.
# So we extract the value from the column name instead.

# Reload normally (as you already did)
solar_cost_df = pd.read_csv(solar_cost_file_path)

# Extract value from column name
solar_thermal_value = solar_cost_df.columns[0]

# Convert to numeric (recommended)
solar_thermal_value = float(solar_thermal_value)

print("Extracted Solar Thermal Cap Cost:", solar_thermal_value)
```

```python
# STEP 6: Create or Replace Column in EPC Dataset
# ----------------------------------------------
epc_df["SOLAR_THERMAL_CAP_COST"] = solar_thermal_value

print(epc_df[["SOLAR_THERMAL_CAP_COST"]].head())
```

```python
# STEP 7: Save Updated Dataset
# ----------------------------
output_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_cap_cost.csv"
epc_df.to_csv(output_path, index=False)

print("Updated file saved to:", output_path)
```
