```python
import pandas as pd

# File path
file_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_wall_update.csv"

# Load data
df = pd.read_csv(file_path)

# Constants
FLOOR_HEIGHT = 2.4
ATTIC_FLOOR_HEIGHT = 2 * FLOOR_HEIGHT  # 4.8 m

# Calculate ridge height assuming 45 degree pitch
# rise = run = half the building width
df["ROOF_RIDGE_HEIGHT"] = ATTIC_FLOOR_HEIGHT + (df["WALL_WIDTH"] / 2)

# Save updated file
output_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_ridge_height_update.csv"
df.to_csv(output_path, index=False)

print("Ridge heights calculated and saved to:", output_path)
```
