```python
# ============================================================
# Step 1: Import Required Libraries
# Description:
# Imports pandas for data handling and pathlib for file paths.
# ============================================================

import pandas as pd
from pathlib import Path
```

```python
# ============================================================
# Step 2: Define File Paths
# Description:
# Specifies the location of the EPC dataset and the output CSV
# that will contain the unique parameter values.
# ============================================================

input_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_priority_index_step8.csv"

output_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_unique_parameter_values.csv"
```

```python
# ============================================================
# Step 3: Load Dataset
# Description:
# Reads the EPC dataset into a pandas DataFrame and prints
# basic information to confirm successful loading.
# ============================================================

df = pd.read_csv(input_path)

print("Dataset loaded successfully.")
print("Number of rows:", len(df))
print("Number of columns:", len(df.columns))
```

```python
# ============================================================
# Step 4: Define Parameters to Analyze
# Description:
# Creates a list of the specific EPC parameters for which
# unique values should be extracted.
# ============================================================

parameters = [
"CURRENT_ENERGY_RATING",
"POTENTIAL_ENERGY_RATING",
"CURRENT_ENVIRONMENTAL_RATING",
"POTENTIAL_ENVIRONMENTAL_RATING",
"WALL_DESCRIPTION",
"WALL_ENERGY_EFF",
"WALL_ENV_EFF",
"ROOF_DESCRIPTION",
"ROOF_ENERGY_EFF",
"ROOF_ENV_EFF",
"FLOOR_DESCRIPTION",
"FLOOR_ENERGY_EFF",
"FLOOR_ENV_EFF",
"WINDOWS_DESCRIPTION",
"WINDOWS_ENERGY_EFF",
"WINDOWS_ENV_EFF",
"MAINHEAT_DESCRIPTION",
"MAINHEAT_ENERGY_EFF",
"MAINHEAT_ENV_EFF",
"MAINHEATCONT_DESCRIPTION",
"MAINHEATC_ENERGY_EFF",
"MAINHEATC_ENV_EFF",
"SECONDHEAT_DESCRIPTION",
"SHEATING_ENERGY_EFF",
"SHEATING_ENV_EFF",
"HOTWATER_DESCRIPTION",
"HOT_WATER_ENERGY_EFF",
"HOT_WATER_ENV_EFF",
"LIGHTING_DESCRIPTION",
"LIGHTING_ENERGY_EFF",
"LIGHTING_ENV_EFF",
"AIR_TIGHTNESS_DESCRIPTION",
"AIR_TIGHTNESS_ENERGY_EFF",
"AIR_TIGHTNESS_ENV_EFF",
"LZC_ENERGY_SOURCES",
"CONSTRUCTION_AGE_BAND",
"EXTENSION_COUNT",
"FLOOR_LEVEL",
"FLAT_TOP_STOREY",
"GLAZED_AREA",
"NUMBER_HABITABLE_ROOMS",
"HEAT_LOSS_CORRIDOOR",
"NUMBER_HEATED_ROOMS",
"LOCAL_AUTHORITY_LABEL",
"MAINS_GAS_FLAG",
"MAIN_HEATING_CATEGORY",
"MAIN_FUEL",
"MAIN_HEATING_CONTROLS",
"MECHANICAL_VENTILATION",
"ENERGY_TARIFF",
"GLAZED_TYPE",
"SOLAR_WATER_HEATING_FLAG",
"TENURE",
"TRANSACTION_TYPE",
"BUILT_FORM",
"PROPERTY_TYPE"
]

print("Total parameters to analyze:", len(parameters))
```

```python
# ============================================================
# Step 5: Extract Unique Values
# Description:
# For each parameter, the script extracts all unique values.
# If a field contains multiple values separated by "|",
# the values are split and treated individually.
# ============================================================

results = []

for param in parameters:

    if param not in df.columns:
        print(f"Warning: {param} not found in dataset.")
        continue

    series = df[param].dropna().astype(str)

    unique_values = set()

    for value in series:
        parts = [p.strip() for p in value.split("|")]
        unique_values.update(parts)

    unique_values = sorted(unique_values)

    results.append({
        "Parameter": param,
        "Unique_Values": "; ".join(unique_values)
    })

print("Unique value extraction complete.")
```

```python
# ============================================================
# Step 6: Create Results DataFrame
# Description:
# Converts the collected unique values into a DataFrame for
# easier export and review.
# ============================================================

unique_values_df = pd.DataFrame(results)

print(unique_values_df.head())
```

```python
# ============================================================
# Step 7: Export Results to CSV
# Description:
# Saves the parameter and unique value summary to a CSV file.
# ============================================================

unique_values_df.to_csv(output_path, index=False)

print("Unique values CSV exported successfully.")
print("File saved to:", output_path)
```
