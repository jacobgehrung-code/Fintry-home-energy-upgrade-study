```python
import pandas as pd

# File path
file_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_hp_cap_cost.csv"

# Load dataset
df = pd.read_csv(file_path)

# Calculate building height
df["BUILDING_HEIGHT"] = df["FLOOR_HEIGHT"] * df["NUMBER_OF_FLOORS"]

# Function to compute exposed perimeter
def exposed_perimeter(row):
    L = row["WALL_LENGTH"]
    W = row["WALL_WIDTH"]
    
    if row["ARCHETYPE"] == "Detached_2F":
        return 2 * (L + W)  # all four sides
    elif row["ARCHETYPE"] == "SemiDetached_2F":
        return 2 * L + W    # three exposed sides
    else:
        return None

# Apply perimeter calculation
df["EXPOSED_PERIMETER"] = df.apply(exposed_perimeter, axis=1)

# Calculate wall area
df["WALL_AREA"] = df["EXPOSED_PERIMETER"] * df["BUILDING_HEIGHT"]

# Save updated dataset
output_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_wall_area_update_2.csv"
df.to_csv(output_path, index=False)

print("Wall area calculation complete. File saved to:", output_path)
```
