```python
import pandas as pd
import numpy as np

# Step 1: Load the dataset
# Description: Load your processed archetype CSV into a pandas DataFrame
file_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_archetypes_extension_update.csv"
df = pd.read_csv(file_path)

```

```python
# Step 2: Define function to calculate wall lengths based on archetype
# Description: Compute WALL_LENGTH and WALL_WIDTH from FOOTPRINT_AREA using aspect ratios
def calculate_walls(row):
    footprint = row["FOOTPRINT_AREA"]  # use individual footprint area per house
    archetype = row["ARCHETYPE"]

    # Semi-Detached (1:1 aspect ratio)
    if "SemiDetached" in archetype:
        length = width = np.sqrt(footprint)
    
    # Detached (3:2 aspect ratio)
    elif "Detached" in archetype:
        # length / width = 3 / 2, footprint = length * width
        # Solve: width = sqrt(footprint * (2/3)), length = 1.5 * width
        width = np.sqrt(footprint * (2 / 3))
        length = 1.5 * width
    
    else:
        # Default fallback if archetype not recognized
        length = width = np.sqrt(footprint)
    
    return pd.Series({"WALL_LENGTH": length, "WALL_WIDTH": width})

```

```python
# Step 3: Apply function to DataFrame
# Description: Add WALL_LENGTH and WALL_WIDTH columns based on archetype
df[["WALL_LENGTH", "WALL_WIDTH"]] = df.apply(calculate_walls, axis=1)

```

```python
# Step 4: Save the updated DataFrame
# Description: Save the new DataFrame with wall lengths for future modeling
output_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_wall_update.csv"
df.to_csv(output_path, index=False)

print("Wall lengths added and saved successfully!")
```
