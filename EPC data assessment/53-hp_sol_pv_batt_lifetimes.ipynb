# STEP 1: Overview — Load EPC dataset and apply lifetime values from lookup table

"""
This script:
1. Loads your EPC dataset containing upgrade lifetime columns.
2. Loads a lookup table (lifetime.csv) that maps each upgrade column to a fixed lifetime value.
3. Applies those lifetime values across all rows in the corresponding EPC dataset columns.
4. Saves the updated dataset to a new CSV file.

Assumptions:
- Column names in EPC dataset exactly match the "UPGRADE" column in lifetime.csv.
- Both files are stored locally and accessible via file paths.
"""

```python
# STEP 2: Import required libraries

import pandas as pd
```

```python
# STEP 3: Load datasets

# EPC dataset
epc_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_cap_cost.csv"
epc_df = pd.read_csv(epc_path)

# Lifetime lookup table
lifetime_path = "/Users/jakegehrung/Desktop/data_raw/REFERENCE FILES/lifetime.csv"
lifetime_df = pd.read_csv(lifetime_path)
```

```python
# STEP 4: Convert lifetime lookup table into dictionary

# Creates mapping like: {"HP_LIFETIME": 15, ...}
lifetime_dict = dict(zip(lifetime_df["UPGRADE"], lifetime_df["LIFETIME"]))

lifetime_dict
```

```python
# STEP 5: Apply lifetime values to EPC dataset columns

for col, value in lifetime_dict.items():
    if col in epc_df.columns:
        epc_df[col] = value
    else:
        print(f"Warning: Column '{col}' not found in EPC dataset")
```

```python
# STEP 6: Save updated dataset

output_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_cap_cost_updated.csv"
epc_df.to_csv(output_path, index=False)

print(f"Updated file saved to: {output_path}")
```
