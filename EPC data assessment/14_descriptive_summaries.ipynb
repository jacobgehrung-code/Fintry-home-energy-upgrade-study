```python
# Step 6a: Load Libraries and Data
# Description: Import necessary Python libraries and load the dataset into a pandas DataFrame
import pandas as pd
import numpy as np

# File path for your dataset
file_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_step4_building_characteristics.csv"

# Load the dataset
df = pd.read_csv(file_path)

# Quick check
df.head()
```

```python
# Step 6b (fixed): Numeric Summaries
# Description: Calculate mean, median, min, max, quartiles, and totals for key numeric fields that exist in the dataset

# Only include columns that exist
numeric_cols = [
    "TOTAL_FLOOR_AREA",
    "ENERGY_CONSUMPTION_CURRENT",
    "3_YR_ENERGY_COST_CURRENT",
    "CO2_EMISS_CURR_PER_FLOOR_AREA",
    "SPACE_HEATING_DEMAND",
    "WATER_HEATING_DEMAND",
    "3_YR_ENERGY_SAVINGS_POTENTIAL",
    "DELTA_ENERGY_EFFICIENCY",
    "DELTA_ENVIRONMENT_IMPACT",
    "DELTA_ENERGY_CONSUMPTION",
    "DELTA_CO2_EMISSIONS",
    "ANNUAL_ENERGY_COST_SAVINGS"
]

# Filter numeric columns to only those actually in df
numeric_cols = [col for col in numeric_cols if col in df.columns]

# Summary stats
numeric_summary = df[numeric_cols].describe(percentiles=[0.25, 0.5, 0.75]).T
numeric_summary["total"] = df[numeric_cols].sum()
numeric_summary
```

```python
# Step 6c: Categorical Summaries
# Description: Count frequencies and percentages for key categorical fields
categorical_cols = [
    "CURRENT_ENERGY_RATING",
    "POTENTIAL_ENERGY_RATING",
    "CURRENT_ENVIRONMENTAL_RATING",
    "POTENTIAL_ENVIRONMENTAL_RATING",
    "MAIN_HEATING_CATEGORY",
    "MAIN_FUEL",
    "CONSTRUCTION_AGE_BAND",
    "BUILT_FORM",
    "PROPERTY_TYPE",
    "TENURE"
]

categorical_summary = {}
for col in categorical_cols:
    counts = df[col].value_counts(dropna=False)
    percentages = df[col].value_counts(normalize=True, dropna=False) * 100
    categorical_summary[col] = pd.DataFrame({"count": counts, "percent": percentages})

# Example: display summary for CURRENT_ENERGY_RATING
categorical_summary["CURRENT_ENERGY_RATING"]
```

```python
# Step 6d: Improvement Potential Summaries
# Description: Summarize delta fields to assess energy and CO2 improvement potential
delta_cols = [
    "DELTA_ENERGY_EFFICIENCY",
    "DELTA_ENVIRONMENT_IMPACT",
    "DELTA_ENERGY_CONSUMPTION",
    "DELTA_CO2_EMISSIONS",
    "ANNUAL_ENERGY_COST_SAVINGS",
]

delta_summary = df[delta_cols].describe(percentiles=[0.25, 0.5, 0.75]).T
delta_summary["total"] = df[delta_cols].sum()
delta_summary
```
