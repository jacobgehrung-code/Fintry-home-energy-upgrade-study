```python
import pandas as pd

# Load dataset
file_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_NPV_IRR1.csv"
df = pd.read_csv(file_path)

# Preview dataset
df.head()
```

```python
df["BUILT_FORM"].value_counts()
```

```python
def assign_archetype(row):
    built_form = str(row["BUILT_FORM"]).strip().lower()
    floors = row["NUMBER_OF_FLOORS"]

    # Semi-detached houses (check first to avoid catching "detached" in this string)
    if "semi-detached" == built_form:
        return f"SemiDetached_{floors}F"

    # Detached houses
    if "detached" == built_form:
        return f"Detached_{floors}F"

    # End terrace
    if "end-terrace" == built_form or "enclosed end-terrace" == built_form:
        return f"EndTerrace_{floors}F"

    # Mid terrace
    if "mid-terrace" == built_form:
        return f"MidTerrace_{floors}F"

    # Flats / apartments
    if "flat" in built_form:
        return f"Flat_{floors}F"

    # Fallback category
    return f"Other_{floors}F"

df["ARCHETYPE"] = df.apply(assign_archetype, axis=1)
```
