# STEP 1 — Overview (Markdown Description)
# This script:
# 1. Loads your EPC dataset from the specified CSV file
# 2. Rounds the LATITUDE and LONGITUDE columns to 4 decimal places
# 3. Saves a new CSV file with the updated values to your specified output path


```python
# STEP 2 — Import Required Libraries
# Load pandas for data manipulation

import pandas as pd
```

```python
# STEP 3 — Define File Paths
# Set input and output file locations

input_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_cons_update6.csv"
output_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_lat_long_4_dec.csv"
```

```python
# STEP 4 — Load Dataset
# Read the CSV file into a pandas DataFrame

df = pd.read_csv(input_path)

# Optional: preview data to confirm load
df.head()
```

```python
# STEP 5 — Round Latitude and Longitude Columns
# Round values to 4 decimal places

df["LATITUDE"] = df["LATITUDE"].round(4)
df["LONGITUDE"] = df["LONGITUDE"].round(4)
```

```python
# STEP 6 — Save Updated Dataset
# Export the modified DataFrame to a new CSV file

df.to_csv(output_path, index=False)

print(f"File successfully saved to: {output_path}")
```
