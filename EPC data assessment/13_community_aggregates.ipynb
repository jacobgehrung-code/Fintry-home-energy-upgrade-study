```python
# Step 5A: Load Libraries and Dataset
# Description: Import necessary Python libraries and read the processed EPC dataset
import pandas as pd
import numpy as np

# File path for your dataset
file_path = '/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_step4_building_characteristics.csv'

# Read CSV into DataFrame
df = pd.read_csv(file_path)

# Quick check of data
df.head()
```

```python
# Step 5B: Select Key Columns for Aggregation
# Description: Define the numeric fields relevant to community-level aggregates
numeric_cols = [
    'ENERGY_CONSUMPTION_CURRENT', 
    'CO2_EMISSIONS_CURRENT', 
    '3_YR_ENERGY_COST_CURRENT',
    '3_YR_ENERGY_SAVINGS_POTENTIAL',
    'ENERGY_CONSUMPTION_POTENTIAL',
    'CO2_EMISSIONS_POTENTIAL',
    'HEATING_COST_CURRENT',
    'HEATING_COST_POTENTIAL',
    'HOT_WATER_COST_CURRENT',
    'HOT_WATER_COST_POTENTIAL',
    'LIGHTING_COST_CURRENT',
    'LIGHTING_COST_POTENTIAL'
]

# Optional categorical columns for breakdowns
categorical_cols = [
    'PROPERTY_TYPE', 
    'CONSTRUCTION_AGE_BAND'
]

# Check selected columns
df[numeric_cols + categorical_cols].head()
```

```python
# Step 5C: Community-Level Totals
# Description: Compute total energy, total CO2, total cost, and total potential savings for the community
community_totals = df[numeric_cols].sum()
community_totals
```

```python
# Step 5D: Community-Level Averages
# Description: Compute average per dwelling for energy, CO2, and cost
community_averages = df[numeric_cols].mean()
community_averages
```

```python
# Step 5E: Aggregates by Dwelling Type
# Description: Compute sums and averages grouped by PROPERTY_TYPE
aggregates_by_type = df.groupby('PROPERTY_TYPE')[numeric_cols].agg(['sum','mean'])
aggregates_by_type
```

```python
# Step 5F: Aggregates by Construction Age Band
# Description: Compute sums and averages grouped by CONSTRUCTION_AGE_BAND
aggregates_by_age = df.groupby('CONSTRUCTION_AGE_BAND')[numeric_cols].agg(['sum','mean'])
aggregates_by_age
```

```python
# Step 5G: Optional – Combine Totals and Averages into a Summary Table
# Description: Create a clean DataFrame showing totals and averages for reporting
summary_table = pd.DataFrame({
    'Total': community_totals,
    'Average per Home': community_averages
})

summary_table
```

```python
# Step 5H: Optional – Export Summary to CSV
# Description: Save the community-level summary to a CSV for FDT website or Excel handover
summary_table.to_csv('/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_community_aggregates_step5.csv')
```
