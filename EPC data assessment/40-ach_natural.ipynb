# STEP 1: DESCRIPTION
# This script loads an EPC dataset, calculates a new column called ACH_NATURAL
# using the formula:
# ACH_NATURAL = [(0.135 * (SPACE_HEATING_DEMAND / TOTAL_FLOOR_AREA)) - 1.17] / 18
# It then saves the updated dataset to a new CSV file.

# No code execution in this step

```python
# STEP 2: IMPORT LIBRARIES
# Import pandas for data handling

import pandas as pd
```

```python
# STEP 3: LOAD DATA
# Load the CSV file into a pandas DataFrame

input_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_orientation.csv"
df = pd.read_csv(input_path)

# Preview the data
df.head()
```

```python
# STEP 4: VALIDATE REQUIRED COLUMNS
# Ensure required columns exist in the dataset

required_columns = ["SPACE_HEATING_DEMAND", "TOTAL_FLOOR_AREA"]

missing_columns = [col for col in required_columns if col not in df.columns]

if missing_columns:
    raise ValueError(f"Missing required columns: {missing_columns}")
else:
    print("All required columns are present.")
```

```python
# STEP 5: CALCULATE ACH_NATURAL (UPDATED WITH ROUNDING)
# Apply the formula and round to the nearest hundredth (2 decimal places)

df["ACH_NATURAL"] = (
    (0.135 * (df["SPACE_HEATING_DEMAND"] / df["TOTAL_FLOOR_AREA"]) - 1.17) / 18
).round(2)

# Preview results
df[["SPACE_HEATING_DEMAND", "TOTAL_FLOOR_AREA", "ACH_NATURAL"]].head()
```

```python
# STEP 6: HANDLE POTENTIAL ISSUES (OPTIONAL CLEANING)
# Replace infinite values or NaNs (e.g., if TOTAL_FLOOR_AREA = 0)

df["ACH_NATURAL"] = df["ACH_NATURAL"].replace([float("inf"), -float("inf")], pd.NA)

# Optional: drop rows with missing ACH_NATURAL
# df = df.dropna(subset=["ACH_NATURAL"])

df.head()
```

```python
# STEP 7: SAVE OUTPUT FILE
# Save the updated dataset to a new CSV

output_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_ach_natural.csv"
df.to_csv(output_path, index=False)

print(f"File saved to: {output_path}")
```
