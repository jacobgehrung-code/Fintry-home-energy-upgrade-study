```python
# Step 1: Load Dataset
# Description: Import required libraries and load the EPC dataset from the specified file path.

"""
This script assigns a WALL_UPGRADE_TYPE to each dwelling in the EPC dataset
based on its WALL_CONS value.

Steps:
1. Load the dataset
2. Define mapping logic (WALL_CONS → WALL_UPGRADE_TYPE)
3. Apply mapping to create/update WALLS_UPGRADE_TYPE column
4. Check for unmapped values
5. Save updated dataset
"""

import pandas as pd

# File path
file_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_hp_peak_load.csv"

# Load dataset
df = pd.read_csv(file_path)

# Preview
df.head()
```

```python
# Step 2: Define Mapping Dictionary
# Description: Create a dictionary that maps WALL_CONS values to WALL_UPGRADE_TYPE.

wall_mapping = {
    "avg_u0.12": "none",
    "avg_u0.13": "none",
    "avg_u0.21": "none",
    "brick_as_built_bands_A_to_C": "external",
    "cavity_as_built_band_G": "external",
    "cavity_as_built_bands_A_to_E": "external_and_cavity",
    "cavity_ext_insul_band_D": "cavity",
    "cavity_filled_band_F": "external",
    "cavity_filled_band_G": "external",
    "cavity_filled_bands_A_to_E": "external",
    "gran_whin_as_built_band_A": "internal",
    "gran_whin_int_insul_band_A": "internal",
    "gran_whin_int_insul_band_G": "internal",
    "park_home_as_built_band_G": "external",
    "park_home_as_built_band_I": "external",
    "park_home_ext_insul_band_F": "external",
    "park_home_ext_insul_band_G": "external",
    "park_home_ext_insul_band_I": "external",
    "park_home_ext_insul_band_K": "external",
    "sand_lime_as_built_band_A": "internal",
    "sand_lime_as_built_band_C": "internal",
    "sand_lime_as_built_band_G": "internal",
    "sand_lime_int_insul_band_J": "internal",
    "system_as_built_band_D": "external",
    "system_as_built_insulated": "internal",
    "system_ext_insul_band_D": "internal",
    "timber_as_built_band_D": "external",
    "timber_as_built_band_F": "external",
    "timber_as_built_band_J": "external",
    "timber_as_built_band_K": "external",
    "timber_as_built_bands_G_to_I": "external"
}
```

```python
# Step 3: Apply Mapping
# Description: Use the mapping dictionary to populate the WALL_UPGRADE_TYPE column.

df["WALLS_UPGRADE_TYPE"] = df["WALL_CONS"].map(wall_mapping)

# Preview results
df[["WALL_CONS", "WALLS_UPGRADE_TYPE"]].head(10)
```

```python
# Step 4: Check for Unmapped Values
# Description: Identify any WALL_CONS values that were not found in the mapping.

unmapped = df[df["WALLS_UPGRADE_TYPE"].isna()]["WALL_CONS"].unique()

print("Unmapped WALL_CONS values:")
print(unmapped)
```

```python
# Step 5: Save Updated Dataset
# Description: Save the updated dataset with WALL_UPGRADE_TYPE filled.

output_path = "/Users/jakegehrung/Desktop/data_raw/data_processed/fintry_epc_wall_upgrade_types.csv"

df.to_csv(output_path, index=False)

print(f"Updated dataset saved to: {output_path}")
```
